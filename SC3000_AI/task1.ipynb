```python
import pandas as pd
import json
import os
import heapq
```

```python
file_name= ['G.json', 'Dist.json']
folder_path = './data'

def load_as_dict(filename):
    path = os.path.join(folder_path, filename)
    with open(path, 'r', encoding='utf-8') as f:
        return json.load(f)

G, Dist = [load_as_dict(f) for f in file_name]
```

```python
def dijkstra(G, Dist, src):
    # Init Vertices to INF
    distances = {node: float('inf') for node in G}
    # Set starting node distance to 0
    distances[src] = 0

    # Add starting node to min heap
    q = [(0, src)]

    # Path to goal node (50)
    predecessor = {node: None for node in G}

    while q:
        current_dist, current_node = heapq.heappop(q)
        
        if current_dist > distances[current_node]:
            continue
        
        # Explore neighbors
        for neighbor in G[current_node]:
            edge_key = f"{current_node},{neighbor}"
            weight = Dist[edge_key]
            
            new_dist = current_dist + weight
            
            if new_dist < distances[neighbor]:
                distances[neighbor] = new_dist
                predecessor[neighbor] = current_node
                heapq.heappush(q, (new_dist, neighbor))
    
    return distances, predecessor
```

```python
distances, predecessor = dijkstra(G, Dist, "1")
```

```python
def get_path(predecessor, src, target):
    path = []
    current = target
    
    while current is not None:
        path.append(current)
        current = predecessor[current]
    
    path.reverse()
    
    # Optional: check if reachable
    if path[0] != src:
        return None  # no path exists
    
    return path
```

```python
distances, predecessor = dijkstra(G, Dist, "1")

path = get_path(predecessor, "1", "264346")

print("Shortest distance:", distances["264346"])
print("Full path:", path)
```
